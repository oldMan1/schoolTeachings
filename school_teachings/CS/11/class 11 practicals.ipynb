Generate the following patterns using nested loops:

![pattern_printing.png](pictures_for_notebook/pattern_printing.png)

```python
# Pattern 1

lines = int(input("Enter number of lines: "))

for line in range(1, lines + 1):
    print('*' * line)
```

```python
# Pattern 2

lines = int(input("Enter number of lines: "))

while lines > 0:
    for num in range(1, lines + 1):
        print(num, end='')
    print()
    lines -= 1
```

```python
# Pattern 3

lines = int(input("Enter number of lines: "))

for line in range(1, lines+1):# 1 to lines
    for ascii in range(line): # 0 to line-1
        print(chr(65 + ascii), end="") # ASCII value of 'A' is 65
    print()
```

```python
# 3rd summation

x = float(input("Enter the value of x: "))
exp = int(input("Enter the value of n: "))

summation = 0
for i in range(1, exp + 1):
    summation += (x ** i) / (i)

print("The sum of the series is:", summation)
```

```python
# 4th summation

from math import factorial

x = float(input("Enter the value of x: "))
exp = int(input("Enter the value of n: "))

summation = 0
for i in range(1, exp + 1):
    summation += (x ** i) / factorial(i)

print("The sum of the series is:", summation)
```

#  Input a number and check if the number is a prime or composite number.

```python
# â— Input a number and check if the number is a prime or composite number.

number = int(input("Enter a number: "))
is_prime = True 
if number <= 1:
    is_prime = False
else:
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
        
if is_prime:
    print(number, "is a prime number.")
else:
    print(number, "is a composite number.")

```
